```python
import pandas as pd
from rdkit import Chem
from rdkit.Chem import Descriptors

df = pd.read_csv('w4-1_rdkit.csv')

desc_total = []

for index, row in df.iterrows():
    smi = row['SMILES']
    
    if pd.isna(smi):   # 🔥 핵심
        desc_total.append([None]*len(Descriptors._descList))
        continue
    
    mol = Chem.MolFromSmiles(smi)

    if mol is not None:
        desc = Descriptors.CalcMolDescriptors(mol)
    else:
        desc = [None]*len(Descriptors._descList)

    desc_total.append(desc)
```

```python
import numpy as np
from rdkit.Chem import AllChem

fpgen = AllChem.GetRDKitFPGenerator()

fp_total = []
for index, row in df.iterrows():
    smi = row['SMILES']
    mol = Chem.MolFromSmiles(smi)
    fp = fpgen.GetFingerprint(mol)
    fp_array = np.zeros((0,), dtype=int)
    Chem.DataStructs.ConvertToNumpyArray(fp, fp_array)
    fp_total.append(fp_array)
```

```python
df_fp = pd.DataFrame(fp_total, columns=[f'bit_{i}' for i in range(len(fp_total[0]))])
```

```python
# smiles 코드가 단일 구조가 아니라면 '.'이 포함되어 있음.
# 단일 구조가 아닌 물질들만 리스트를 확인하고, 제거해야 될 것과 제거하면 안되는 물질 구분해보기.
# 현재 대부분이 유기 화합물이기 때문에, 유기화합물이 염(salt)가 추가된 경우는, 염을 제거한 smiles 코드를 남겨야함.
# 무기화합물이라면 제거. 유효 성분이 몇개인지 확인이 어려운 경우도 제거.
```

```python
df_multi = df[df['SMILES'].str.contains(r'\.', na=False)]
print(len(df_multi))
df_multi[['Chemical_Name', 'SMILES']].head(20)
def split_smiles(smi):
    if pd.isna(smi):
        return []
    return smi.split('.')

df_multi['components'] = df_multi['SMILES'].apply(split_smiles)

df_multi[['SMILES', 'components']].head(10)
df_multi['num_components'] = df_multi['components'].apply(len)

df_multi[['SMILES', 'num_components']].head()
```

```python
from rdkit import RDLogger
RDLogger.DisableLog('rdApp.*')

def refine_smiles(smi):
    if pd.isna(smi):
        return None
    
    parts = smi.split('.')
    organic_parts = []
    
    for part in parts:
        mol = Chem.MolFromSmiles(part)
        if mol is None:
            continue
        
        # 유기화합물 조건: 탄소 포함
        if any(atom.GetSymbol() == 'C' for atom in mol.GetAtoms()):
            organic_parts.append(mol)
    
    # ❌ 유기물 없음 → 제거
    if len(organic_parts) == 0:
        return None
    
    # ❌ 유기물이 2개 이상 → 제거
    if len(organic_parts) > 1:
        return None
    
    # ✅ 유기물 1개만 남기기
    return Chem.MolToSmiles(organic_parts[0])


# =========================
# ✅ 적용
# =========================
df['refined_SMILES'] = df['SMILES'].apply(refine_smiles)

# 유효한 데이터만 남기기
df_clean = df.dropna(subset=['refined_SMILES']).reset_index(drop=True)
```

```python
from rdkit import Chem
import pandas as pd

def is_organic(mol):
    # 탄소(C)가 하나라도 있으면 유기화합물로 판단
    return any(atom.GetAtomicNum() == 6 for atom in mol.GetAtoms())

def refine_smiles(smi):
    if pd.isna(smi):
        return None

    parts = smi.split('.')
    organic_parts = []

    for part in parts:
        mol = Chem.MolFromSmiles(part)
        if mol is None:
            continue

        if is_organic(mol):
            organic_parts.append((part, mol.GetNumHeavyAtoms()))

    # 유기 성분이 딱 1개면 → 그거 사용
    if len(organic_parts) == 1:
        return organic_parts[0][0]

    # 유기 성분이 여러 개면 → 가장 큰 것만 남기기 (salt 제거)
    elif len(organic_parts) > 1:
        organic_parts.sort(key=lambda x: x[1], reverse=True)
        return organic_parts[0][0]

    # 유기 성분 없으면 → 무기 → 제거
    return None


# 적용
df['refined_smiles'] = df['smiles'].apply(refine_smiles)

# None 제거
df_clean = df.dropna(subset=['refined_smiles'])

print(len(df_clean))
```
