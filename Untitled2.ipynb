```python
import pandas as pd
from sklearn.feature_selection import VarianceThreshold, SelectKBest, f_classif

# 1. 데이터 로드 및 결측치 제거
df = pd.read_csv('skin_irritation_2Ddesc.csv')
X_raw = df.drop(columns=['Chemical_Name', 'standardized_smi', 'label'])
y = df['label']

# 결측치 행 제거 및 인덱스 동기화
X_clean = X_raw.dropna()
y_clean = y[X_clean.index]

# ---------------------------------------------------------
# [단계 1] 표준편차 0.01 미만 제거
# ---------------------------------------------------------
selector_var = VarianceThreshold(threshold=(0.01**2))
X_high_variance = selector_var.fit_transform(X_clean)
remaining_cols = X_clean.columns[selector_var.get_support()]

# 데이터프레임으로 다시 변환 (X_filtered 생성)
X_filtered = pd.DataFrame(X_high_variance, columns=remaining_cols)

# ---------------------------------------------------------
# [단계 2] SelectKBest를 이용한 특성 선택
# ---------------------------------------------------------
k_value = 20
selector_k = SelectKBest(score_func=f_classif, k=k_value)
X_selected = selector_k.fit_transform(X_filtered, y_clean)

# 최종 선택된 특성 이름 확인
selected_features_names = X_filtered.columns[selector_k.get_support()]

print(f"1단계 후 남은 특성: {X_filtered.shape[1]}개")
print(f"2단계 후 최종 선택된 특성: {X_selected.shape[1]}개")
print(f"최종 특성 리스트: {list(selected_features_names)}")

# 최종 학습용 데이터프레임
X_final = pd.DataFrame(X_selected, columns=selected_features_names)
```

```python
from sklearn.neural_network import MLPClassifier
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

# [데이터 준비] 2번 단계에서 만든 X_final과 y_clean 사용
# MLP는 데이터 스케일에 민감하므로 StandardScaler가 필수입니다.
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X_final)

# 학습/테스트 데이터 분리
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y_clean, test_size=0.2, random_state=42)

# [3단계] MLP 모델 정의 및 하이퍼파라미터 설정
mlp = MLPClassifier(
    hidden_layer_sizes=(64, 32), # 2개의 은닉층
    activation='relu',           # 활성화 함수
    solver='lbfgs',              # 소규모 데이터에 적합한 솔버
    alpha=0.01,                  # 규제 강도
    max_iter=1000,               # 충분한 학습 횟수
    random_state=42
)

# 모델 학습
mlp.fit(X_train, y_train)

# 결과 체크
print(f"훈련 세트 정확도: {mlp.score(X_train, y_train):.3f}")
print(f"테스트 세트 정확도: {mlp.score(X_test, y_test):.3f}")
```

```python
# hidden_layer_sizes': [(5,), (10,), (5, 5), (10, 5)]
# (5, ): 은닉층(Hidden Layer)을 1개만 만들고, 그 안에 뉴런(노드)을 5개 배치해라.
# (10, ): 은닉층을 1개만 만들고, 뉴런을 10개 배치해라.
# (5, 5): 은닉층을 2개 만들고, 첫 번째 층에 5개, 두 번째 층에도 5개의 뉴런을 배치해라.
# (10, 5): 은닉층을 2개 만들고, 첫 번째 층에 10개, 두 번째 층에 5개의 뉴런을 배치해라.
```

```python
from sklearn.neural_network import MLPClassifier
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import accuracy_score, classification_report

# 1. MLP 하이퍼파라미터 그리드 설정
# 수진님의 데이터 규모에 맞춰 노드 수와 복잡도를 조절했습니다.
param_grid = {
    'hidden_layer_sizes': [(5,), (10,), (5, 5), (10, 5)], # 층과 노드 구성 (1층, 2층)
    'activation': ['tanh', 'relu'],                      # 활성화 함수
    'solver': ['lbfgs', 'adam'],                         # 소규모 데이터는 lbfgs 권장
    'alpha': [0.0001, 0.01, 0.1],                        # L2 규제 (과적합 방지)
    'max_iter': [1000]                                   # 충분한 학습 횟수
}

# 2. GridSearchCV 설정
# 상위 5개로 변환된 X_train_5 데이터를 사용합니다.
mlp = MLPClassifier(random_state=42)
grid_search = GridSearchCV(mlp, param_grid, cv=5, scoring='accuracy', n_jobs=-1)

# 3. 모델 학습 및 최적 조합 찾기
grid_search.fit(X_train_5, y_train)

# 4. 결과 확인
print("="*50)
print(f"최적 하이퍼파라미터: {grid_search.best_params_}")
print(f"최고 교차 검증 점수: {grid_search.best_score_:.4f}")
print("="*50)

# 5. 최적 모델로 테스트 데이터 평가
best_model = grid_search.best_estimator_
y_pred = best_model.predict(X_test_5)

train_acc = best_model.score(X_train_5, y_train)
test_acc = accuracy_score(y_test, y_pred)

print(f"최적 모델 학습 정확도: {train_acc:.4f}")
print(f"최적 모델 테스트 정확도: {test_acc:.4f}")
print("-"*50)
print("[상세 성능 리포트]")
print(classification_report(y_test, y_pred))
```
